```python
import json

def load_json(filepath):
    dict = {}
    with open(filepath) as jsonfile:
        dict = json.load(jsonfile)
    
    return dict
```

```python
old = load_json('../support/recommendations.json')
new = load_json('test_recommendations.json')
```

```python
len(old.keys()) == len(new.keys())
```

```python
for id in old:
    for rec in old[id]:
        if rec not in new[id]:
            print('fail')
```
