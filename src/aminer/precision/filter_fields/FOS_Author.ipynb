```python
import json

from tqdm import tqdm
```

```python
def load_json(filename):
    d = None
    with open(filename) as file:
        d = json.load(file)
    
    return d
```

```python
fos = load_json('fos_filtered_recommendations.json')
author = load_json('author_filtered_citations.json')
```

```python
both = {}
for id in fos:
    if id in author:
        both[id] = list(set(author[id]).intersection(fos[id]))
    else:
        both[id] = fos[id]
```

```python
sum([len(both[a]) for a in both]) / len(both)
```

```python
def dump_json(filename, d):
    with open(filename, 'w') as file:
        json.dump(d, file)
```

```python
dump_json('fos_author_filtered.json', both)
```

```python
from aminer.recall.query_es import get_abstract_by_pids, get_references_by_pid

test_recs = {}
for id in both:
    test_recs[id] = get_references_by_pid(id)
```

```python
overall = 0
for id in tqdm(both):
    points = 0
    for rec_id in test_recs[id]:
        points += int(rec_id in author[id])
    
    overall += points / len(test_recs[id])
    print(points / len(test_recs[id]))
```
